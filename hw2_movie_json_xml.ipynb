# JSON 방식

```python
import pandas as pd
import urllib.request
import json
```

```python
url_p1 = "http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
key = "f5eef3421c602c6cb7ea224104795888"
url_p2 = "&itemPerPage=50&curPage=5&movieTypeCd=220101"
url = url_p1 + "?key=" + key + url_p2
url
```

```python
movie_page = urllib.request.urlopen(url)
movie_data = json.loads(movie_page.read().decode("utf-8"))
movie_data
```

```python
# 수집할 정보 : 영화 코드값, 영화 제목, 영화 제목 영문명, 상영시간, 심의등급, 1번 배우있다면 이름과 배역에 대한 정보를 수집하세요.
# 250개 영화
# 영화목록 데이터
"movieCd"
"movieNm"
"movieNmEn"
# 영화 상세정보 데이터
"showTm"
"watchGradeNm"
"actors"
```

```python
movie_data["movieListResult"]["movieList"]
```

```python
tot_data = []

for idx, data in enumerate(movie_data['movieListResult']["movieList"]):
  
  i_code = data["movieCd"]
  i_name = data["movieNm"]
  i_Ename = data["movieNmEn"]
  tot_data.append([i_code, i_name, i_Ename])

tot_data

```

```python
movie_df = pd.DataFrame(data = tot_data, columns = ["code","movie_Name", "movie_EnName"])
movie_df
```

```python
len(movie_df["code"].unique()) == len(movie_df)
```

```python
url_base = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"
key = "f5eef3421c602c6cb7ea224104795888"
for i in movie_df.index:
  url = url_base + "?key=" + key + "&movieCd=" + i
  temp_res = urllib.request.urlopen(url)
  temp_data = json.loads(temp_res.read().decode("utf-8"))
  if temp_data["movieInfoResult"]["movieInfo"]["showTm"] == []:
    movie_df.at[i,"showTm"] = "NA"
  else:
    movie_df.at[i,"showTm"] = temp_data["movieInfoResult"]["movieInfo"]["showTm"]
  if temp_data["movieInfoResult"]["movieInfo"]["audits"] == []:
    movie_df.at[i,"watchGrade"] = 'NA'
  else:
    movie_df.at[i, "watchGrade"] = temp_data["movieInfoResult"]["movieInfo"]["audits"][0]["watchGradeNm"]
  if temp_data["movieInfoResult"]["movieInfo"]["actors"] == []:
    movie_df.at[i,"firstActor"] = "NA"
  else:
    movie_df.at[i,"firstActor"] = temp_data["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"]
print("DONE")
```

# XML 방식

```python
from bs4 import BeautifulSoup
```

```python
url_p1 = "http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml"
key = "f5eef3421c602c6cb7ea224104795888"
url_p2 = "&itemPerPage=50&curPage=5&movieTypeCd=220101"
url = url_p1 + "?key=" + key + url_p2
url
```

```python
movie_page = urllib.request.urlopen(url)
soup = BeautifulSoup(movie_page, "html.parser")
```

```python
print(soup)
```

```python
soup.find("movie")
```

```python
# 영화목록 데이터
# "movieCd"
# "movieNm"
# "movieNmEn"
soup.find_all("movie")[0].find("moviecd").text #moviecd
soup.find_all("movie")[0].find("movienm").text #movienm
soup.find_all("movie")[0].find("movienmen").text #movienmen
```

```python
from collections import deque
```

```python
movie_list = deque([])

tot_cnt = len(soup.find_all("movie"))

for idx, data in enumerate(soup.find_all("movie")):
  code = data.find("moviecd").text
  name = data.find("movienm").text
  enName = data.find("movienmen").text
  movie_list.append([code, name, enName]) 

print("Done")
  
```

```python
movie_list
```

```python
movie_df = pd.DataFrame(movie_list, 
             columns = ["code","movieName","EnMovieName"])
```

```python
for i in movie_df.index:
  url_base = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml"
  key = "f5eef3421c602c6cb7ea224104795888"
  url = url_base + "?key=" + key + "&movieCd=" + i
  temp_res = urllib.request.urlopen(url)
  soup2 = BeautifulSoup(temp_res,"html.parser")
  if soup2.find("movieinforesult").find("movieinfo").find("showtm")==[]:
    movie_df.at[i,"showTm"] = "NA"
  else:
    movie_df.at[i,"showTm"] = soup2.find("movieinforesult").find("movieinfo").find("showtm").text
  if soup2.find("movieinforesult").find("movieinfo").find("audits").find("watchgradenm") == []:
    movie_df.at[i,"watchGrade"] = "NA"
  else:
    movie_df.at[i,"watchGrade"] = soup2.find("movieinforesult").find("movieinfo").find("audits").find("watchgradenm")
  if soup2.find("movieinforesult").find("movieinfo").find_all("actors") == []:
    movie_df.at[i,"actors"] = "NA"
  else:
    movie_df.at[i,"actors"] = soup2.find("movieinforesult").find("movieinfo").find_all("actors")[0].find("peoplenm") 

print("Done")
```
